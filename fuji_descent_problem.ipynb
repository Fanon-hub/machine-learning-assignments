```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

csv_path='mtfuji_data.csv'
np.set_printoptions(suppress=True)
fuji = np.loadtxt(csv_path, delimiter=',', skiprows=1)
print(fuji[130:140])

file_path = 'mtfuji_data.csv'
fuji_data = pd.read_csv(file_path)

# extracting point numbers (x) and elevation 
x_points = fuji_data["x"]
elevation = fuji_data["elevation"]
# plotting the elevation data
plt.figure(figsize=(10, 6))
plt.plot(x_points, elevation, marker='*', label='Elevation Profile', color='skyblue',linestyle='-')
plt.xlabel('Point Number')
plt.ylabel('Elevation (m)')
plt.title('Elevation Profile of Mt. Fuji')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
x = fuji[:,0]
y = fuji[:,3]

def gradient(m):
    if m == 0:
        return 0
    return (y[m] - y[m-1]) / (x[m] - x[m-1])
gradient(100)
```

```python
def move(m, alpha):
    new_index = fuji[m, 3] - alpha * gradient(m)
    new_index = max(0, round(new_index))
    return new_index

move(299,0.2)
```

```python
def down(m):
    point = []
    for k in range(299 - m):
        point.append(move (m + k, 0.2))
    return point

down(136)

```

```python
import matplotlib.pyplot as plt
plt.title("Mt.Fuji from No. 136")
plt.xlabel("position")
plt.ylabel("elevation[m]")
plt.tick_params(labelsize=15)
plt.plot(down(136))

plt.show()
```
