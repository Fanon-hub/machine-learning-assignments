```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-50, 50.1, 0.1)  

y=(1/2) * x+1

# calculating the gradient using the differential formula 
dx = np.diff(x)
dy = np.diff(y)
gradient = dy/dx

# printing the results
print("Gradient array:")
print(gradient)
print("\nNumber of gradients:", len(gradient))

# plotting the linear function 
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1) #for the first graph
plt.plot(x, y, label='y=(1/2)x+1', color='orange')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, linestyle='-') #add y-axis line
plt.axvline(0, color='black',linewidth=0.5, linestyle='-') #add x-axis line
plt.grid()
plt.legend()

# plotting the gradient
x_midpoints = (x[:-1] + x[1:]) / 2 #computes the midpoints of x for plotting
plt.subplot(1, 2, 2) #for the second graph
plt.plot(x_midpoints, gradient, label='Gradient', color='magenta')
plt.title('Gradient of Linear Function')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.axhline(0, color='black',linewidth=0.5, linestyle='-') #add y-axis line
plt.grid()
plt.legend()
# showing the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    # Generate x and y values
    x = np.arange(*x_range)
    y = function(x)
    # Combine x and y into a single array
    array_xy = np.column_stack((x, y))
    # Compute the gradient
    dx = np.diff(x)
    dy = np.diff(y)
    gradient = dy / dx
    return array_xy, gradient

# Define functions for the equations
def function1(array_x):
    return array_x**2

def function2(array_x):
    return 2 * array_x ** 2 + 2 * array_x

def function3(array_x):
    return np.sin(array_x / 12)

# Compute gradients for the functions
x_range = (-50, 50.1, 0.1)
array_xy1, gradient1 = compute_gradient(function1, x_range)
array_xy2, gradient2 = compute_gradient(function2, x_range)
array_xy3, gradient3 = compute_gradient(function3, x_range)

# Plotting function and gradient
def plot_function_and_gradient(array_xy, gradient, title_func, title_grad):
    x = array_xy[:, 0]
    y = array_xy[:, 1]
    x_midpoints = (x[:-1] + x[1:]) / 2

    plt.figure(figsize=(10, 6))
    # Plot the function
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label=title_func, color='orange')
    plt.title(title_func)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.grid()
    plt.legend()

    # Plot the gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_midpoints, gradient, label='Gradient', color='magenta')
    plt.title(title_grad)
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot results for each function
plot_function_and_gradient(array_xy1, gradient1, "y = x^2", "Gradient of y = x^2")
plot_function_and_gradient(array_xy2, gradient2, "y = 2x^2 + 2x", "Gradient of y = 2x^2 + 2x")
plot_function_and_gradient(array_xy3, gradient3, "y = sin(x/12)", "Gradient of y = sin(x/12)")

# output gradients as ndarrays
print("Gradients for y=x^2:", gradient1)
print("Gradients for y=2x^2+2x:", gradient2)
print("Gradients for y=sin(x/12):", gradient3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function definitions
def function1(array_x):
    return array_x**2

# Compute y values
x_range = (-50, 50.1, 0.1)
x_values = np.arange(*x_range)
y_values = function1(x_values)

# Find the minimum value and its index
min_y = y_values.min()
min_index = y_values.argmin()
min_x = x_values[min_index]

# Compute gradient
dx = np.diff(x_values)
dy = np.diff(y_values)
gradient = dy / dx

# Gradients before and after the minimum
if min_index > 0:  # Ensure we don't go out of bounds
    gradient_before = gradient[min_index - 1]
else:
    gradient_before = None  # No gradient before the first point

if min_index < len(gradient):  # Ensure we don't go out of bounds
    gradient_after = gradient[min_index]
else:
    gradient_after = None  # No gradient after the last point

# Display results
print(f"Minimum y-value: {min_y}")
print(f"x-value at minimum y: {min_x}")
print(f"Gradient before minimum y: {gradient_before}")
print(f"Gradient after minimum y: {gradient_after}")

# Plot function and gradient
plt.figure(figsize=(10, 5))

# Plot the function
plt.subplot(1, 2, 1)
plt.plot(x_values, y_values, label='y = x^2', color='orange')
plt.scatter(min_x, min_y, color='red', label=f"Minimum (x={min_x:.1f}, y={min_y:.1f})")
plt.title("Function y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()

# Plot the gradient
x_midpoints = (x_values[:-1] + x_values[1:]) / 2
plt.subplot(1, 2, 2)
plt.plot(x_midpoints, gradient, label="Gradient", color='magenta')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Gradient of y = x^2")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
