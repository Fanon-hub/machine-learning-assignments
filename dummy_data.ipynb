```python
import numpy as np

np.random.seed(0)
# mean and covariance matrix
mean = [-3, 0]
covariance_matrix = [[1.0, 0.8], [0.8, 1.0]]
# first set of 500 random numbers
random_numbers_q1 = np.random.multivariate_normal(mean, covariance_matrix, 500)
# second set of 500 random numbers
random_numbers_q2 = np.random.multivariate_normal(mean, covariance_matrix, 500)
# label 0 for data from first set
labels_q1 = np.zeros((500,1))
# label 1 for data from second set
labels_q2 = np.zeros((500,1))
# combining data with labels from both sets
random_numbers_q1_with_labels=np.hstack((random_numbers_q1, labels_q1))
random_numbers_q2_with_labels=np.hstack((random_numbers_q2, labels_q2))
# combining data from both sets
random_numbers_with_labels=np.vstack((random_numbers_q1_with_labels, random_numbers_q2_with_labels))
# verifying the shape of the combined data
print("Shape of the combined data with labels:", random_numbers_with_labels.shape)
# confirm few rows
print("First 5 rows of the combined data with labels:", random_numbers_with_labels[:5])

```
