```python
# The formula to calculate the total number of grains of wheat on the chessboard is:
# Total grains = 2^n - 1, where n is the total number of squares on the chessboard.
# In a standard chessboard, there are 64 squares.

# Calculating the total number of grains.
total_grains = 2**64 - 1
total_grains
```

```python
# Using the same formula, we calculate the total number of grains on a chessboard with 64 squares.
# Grains are placed in powers of 2: 1, 2, 4, 8,...

# Total number of squares on the chessboard
n = 64

# Calculate the total grains: 2^n - 1
total_grains = 2**n - 1

# Convert grains to weight (assuming 1 grain of wheat weighs approximately 0.065 grams)
grain_weight_grams = 0.065  # weight of one grain in grams

# Calculate the total weight in grams and kilograms
total_weight_grams = total_grains * grain_weight_grams
total_weight_kilograms = total_weight_grams / 1000

total_grains, total_weight_grams, total_weight_kilograms
```

```python
import numpy as np

# Define a 2x2 chessboard as a NumPy array
n_squares = 2
chessboard = np.zeros((n_squares, n_squares), dtype=int)

# Fill the chessboard with wheat counts
# Start with 1 grain in the top-left corner, then double for each subsequent square
wheat_count = 1
for i in range(n_squares):
    for j in range(n_squares):
        chessboard[i, j] = wheat_count
        wheat_count *= 2

# Transform the chessboard array into a flat array 
flat_chessboard = chessboard.flatten()

print("2x2 Chessboard with wheat counts:")
print(chessboard)
print("Flattened Chessboard:")
print(flat_chessboard)
```

```python
import numpy as np

def chessboard_grains (n, m):
    n=int(n) #Number of rows
    m=int(m) #Number of columns

#creating a linear array with powers of 2 for the entire chessboard
    grains = np.array([2**i for i in range(n* m)]) 
    chesboard = grains.reshape((n, m))
    return chesboard

# finding the number of grains on the 8 by 8 chessboard
chessboard_8x8 = chessboard_grains(8, 8)
total_grains_8x8 = np.sum(chessboard_8x8)

print("8x8 chessboard:")
print(chessboard_8x8)
print(f"Total grains on an 8x8 chessboard are: {total_grains_8x8}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# creating a 8x8 chessboard where each square will represent grains of wheat (2^n -1)
rows, cols = 8, 8
grains = np.array([[2**(row*cols +col) for col in range(cols)] for row in range(rows)])

total_grains = np.sum(grains)
column_averages = np.mean(grains, axis=0)

print(f"Total grains of wheat on the 8x8 chessboard are: {total_grains}")
print(f"Average grains of wheat per column on the 8x8 matrix chessboard are: {column_averages}")

# plotting the averages on a bar graph 
plt.figure(figsize=(8, 6))
plt.bar(range(1, cols+1), column_averages, color='magenta', linestyle='-', edgecolor='beige')
plt.title('Average Grains of wheat per column on a 8x8 matrix chessboard', fontsize=16)
plt.xlabel('Column', fontsize=14)
plt.ylabel('Average Grains', fontsize=14)
plt.xticks(range(1, cols+1))
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.tight_layout()

total_grains, plt.show()    

```

```python
import numpy as np

def chessboard_grains_loop(n,m):
    n=int(n)
    m=int(m)
    grains = np.array([2**i for i in range(n*m)])
    return grains.reshape(n,m)
# Create an 8 x 8 chessboard
chessboard = chessboard_grains_loop(8,8)
# calculating grains in the first and second halves of the chessboard
first_half_sum = np.sum(chessboard[:4,:].sum(axis=1))
second_half_sum = np.sum(chessboard[4:,:].sum(axis=1))
# multiply the sums
result = first_half_sum * second_half_sum
print(f"The Sum of wheat count grains in the first half is: {first_half_sum},")
print(f"The Sum of the wheat count grains in the second half is:{second_half_sum}.")
print(f"The total result of the two counts is {result}")
```

```python
import numpy as np

def chessboard_grains_append(n,m):
    n=int(n)
    m=int(m)
    grains = np.array([1]) #Starting with the first square having a single grain
    for i in range(1,n*m):
        grains = np.append(grains,grains[-1]*2) #append double the last value
    return grains.reshape(n,m)

def chessboard_grains_broadcast(n,m):
    n=int(n)
    m=int(m)
    # using broadcasting to compute powers of 2
    powers = np.arange(n*m).reshape(n,m)
    grains = 2**powers
    return grains
# Calculating grains for a 8x8 chessboard using both functions
chessboard_8x8_append = chessboard_grains_append(8,8)
chessboard_8x8_broadcast = chessboard_grains_broadcast(8,8)

# output
print("chessboard generated using np.append():")
print(chessboard_8x8_append)
# output
print("chessboard generated using broadcasting:")
print(chessboard_8x8_broadcast)


# verification for if both methods give the same result
np.allclose(chessboard_8x8_append,chessboard_8x8_broadcast)

```
