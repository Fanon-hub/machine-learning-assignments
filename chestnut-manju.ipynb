```python
import math
import matplotlib.pyplot as plt

# fucntion for calculating the time required
def time_required(initial_volume, target_volume, doubling_time=1):

    if initial_volume >= target_volume:
        return 0 #already fillled
    time_minutes = math.log2(target_volume / initial_volume) * doubling_time

    return time_minutes
# function to calculate simulation growth

def simulate_growth(initial_volume, target_volume, doubling_time, max_time):
    times = []
    volumes = []
    current_volume = initial_volume
    time = 0

    while current_volume < target_volume:
        if time > max_time:
            break
        times.append(time)
        volumes.append(current_volume) 
        current_volume *= 2
        time += doubling_time
        

    return time, volumes

# chestnut bun volume (approx. 10 cm radius)
initial_volume = 4/3 * math.pi * (0.1**3) # in cubic meters
# solar system volume (approx. radius to jupiter)
solar_system_radius = 7.785e11 # in meters
solar_system_volume = 4/3 * math.pi * (solar_system_radius**3) # in cubic meters
tokyo_dome_volume = 1239000 # in cubic meters
# doubling time of a chestnut bun in minutes
doubling_time = 1

# calculating the time required to fill the solar system
time_required_solar_system = time_required(initial_volume, solar_system_volume, doubling_time)
# calculating the time required to fill the Tokyo Dome
time_required_tokyo_dome = time_required(initial_volume, tokyo_dome_volume, doubling_time)

print(f"Time required to fill the Tokyo Dome with chestnut buns: {time_required_tokyo_dome / 60 / 24:.2f} days")
print(f"Time required to fill the solar system with chestnut buns: {time_required_solar_system / 60 / 24:.2f} days")

max_time=100 #maximum time in the simulation 
# simulating the growth of chestnut buns in the solar system 
times_solar, volumes_solar, = simulate_growth(initial_volume, solar_system_volume, doubling_time, max_time)
# simulating the growth of chestnut buns in the solar system
times_tokyo, volumes_tokyo, = simulate_growth(initial_volume, tokyo_dome_volume, doubling_time, max_time)



import matplotlib.pyplot as plt

# Example data (replace with your actual data)
times_solar = range(101)  # Example: 101 time steps for solar system
volumes_solar = [2 ** t for t in times_solar]  # Example: Exponential growth
times_tokyo = range(101)  # Example: 101 time steps for Tokyo Dome
volumes_tokyo = [2 ** t for t in times_tokyo]  # Example: Exponential growth

# Create a figure with subplots
plt.figure(figsize=(14, 6))  # Single figure for both subplots

# Plotting the growth of chestnut buns in the solar system
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(times_solar, volumes_solar, color='green', linestyle=':', linewidth=2, label='Chestnut Bun Volume')
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")
plt.title("Exponential Growth in the Solar System", fontsize=14)
plt.yscale("log")  # Logarithmic scale for volume
plt.legend()
plt.grid(True)

# Plotting the growth of chestnut buns in the Tokyo Dome
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(times_tokyo, volumes_tokyo, color='blue', linestyle='-', linewidth=2, label='Chestnut Bun Volume')
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")
plt.title("Exponential Growth in the Tokyo Dome", fontsize=14)
plt.yscale("log")  # Logarithmic scale for volume
plt.legend()
plt.grid(True)

# Adjust layout to avoid overlapping elements
plt.tight_layout()

# Show the plots
plt.show()
```
